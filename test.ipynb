```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('training.csv')
images = data['Image']
images = images.apply(lambda x: np.fromstring(x, sep=' '))

plt.imshow(np.array(data['Image'][1].split()).astype(np.uint8).reshape(96,96), cmap='gray')
# visualize 6 images in a 2x3 grid
fig, axarr = plt.subplots(2, 3)
for i in range(2):
    for j in range(3):
        axarr[i, j].imshow(images[3*i+j].reshape(96,96), cmap='gray')
        axarr[i, j].set_xticks([])
        axarr[i, j].set_yticks([])
        axarr[i, j].set_title('Image %d' % (3*i+j+1))
plt.show()
```
